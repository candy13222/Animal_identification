```python
import cv2
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.applications.densenet import DenseNet169
import matplotlib.pyplot as plt
from PIL import Image

num_classes = 4

model = Sequential()
model.add(DenseNet169(input_shape=(32, 32, 3), include_top=False, weights='imagenet', pooling='avg'))
model.add(Dense(num_classes, activation='softmax'))

model.load_weights("/content/gdrive/MyDrive/ColabNotebooks/weight.h5")
    
img = cv2.imread("/content/gdrive/MyDrive/ColabNotebooks/deer.jpg")
img = cv2.resize(img, (32,32))
imgg = np.zeros((1,32,32,3), dtype="float32")
imgg[0] = img

cl = ["bird", "deer", "dog", "frog"]

ans = model.predict(imgg)[0]

im = Image.open("/content/gdrive/MyDrive/ColabNotebooks/deer.jpg")
plt.subplot(121)
plt.imshow(im)
plt.subplot(122)
plt.bar(cl, ans)
plt.show()
plt.imshow(img)
```
